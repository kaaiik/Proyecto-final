# Tratamiento de cada dataframe

EL objetivo es hacer un plot parecido al que se hace en el documento de 'sea_level.ipynb', por lo tanto, vamos a obtener la media de cada mes para cada ciudad y agruparlas en total por años.

```python
# Con esta agrupación vamos a obtener la suma total

meses2005 = df2005.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2005['Año'] = [2005]*len(meses2005)
meses2005.head()
```

```python
meses2006 = df2006.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2006['Año'] = [2006]*len(meses2006)
meses2006.head()
```

```python
meses2007 = df2007.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2007['Año'] = [2007]*len(meses2007)
meses2007.head()
```

```python
meses2008 = df2008.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2008['Año'] = [2008]*len(meses2008)
meses2008.head()
```

```python
meses2009 = df2009.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2009['Año'] = [2009]*len(meses2009)
meses2009.head()
```

```python
meses2010 = df2010.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2010['Año'] = [2010]*len(meses2009)
meses2010.head()
```

```python
meses2011 = df2011.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2011['Año'] = [2011]*len(meses2005)
meses2011.head()
```

```python
meses2012 = df2012.resample(rule='M', on='Date')['Tanchau_fluxe', 'Chaudocfluxes', 'Cantho_fluxe'].sum()
meses2012['Año'] = [2012]*len(meses2005)
meses2012.head()
```

```python
fluxes = pd.concat([meses2005, meses2006, meses2007, meses2008, meses2009, meses2010, meses2011, meses2012])
fluxes.head()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

plt.ylim(min(fluxes.Tanchau_fluxe)-1,
         max(fluxes.Tanchau_fluxe)+1)
plt.plot(months, fluxes.loc['2005-01-31':'2005-12-31'].Tanchau_fluxe);
```

```python
%matplotlib notebook
plt.rcParams['animation.embed_limit'] = 2**128
```

```python
fig=plt.figure(figsize=(8,5))
ax=plt.subplot(111)

plt.ylim(min(fluxes.Tanchau_fluxe)-1,
         max(fluxes.Tanchau_fluxe)+1)

def animation_Tanchau(i):
 
    plt.title(f'Año {2005 + i}')

    data_plot=fluxes.Tanchau_fluxe[fluxes.Año==2005 + i]    
    
    ax.plot(months[:len(data_plot)], 
            data_plot, 
            c=plt.cm.viridis(i*50));
    
    return ax

anim=FuncAnimation(fig, 
                   animation_Tanchau, 
                   frames=len(fluxes.Año.unique()),
                   interval=500,
                   repeat=False)

anim.save(r'../images/Tanchau.gif')
```
